```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
df
```

```python
y=df['Survived']
x=df.copy()
x.drop(['PassengerId','Name','Ticket','Cabin','Survived'],axis=1, inplace=True)
```

```python
print(y)
x
```

```python
x.replace('male', 0, inplace=True)
x.replace('female', 1, inplace=True)
x.replace('S',1,inplace=True)
x.replace('C',2,inplace=True)
x.replace('Q',3,inplace=True)
x
```

```python
for i in range(7):
    m=x.iloc[:,i].mean()
    x.iloc[:,i].replace(np.nan,m, inplace=True)
x
```

```python
columns=[]
for col in x:
    columns.append(col)
columns
```

```python
for i in range(7):
    col=columns[i]
    mn=x[col].min()
    mx=x[col].max()
    x[col]=(x[col]-mn)/(mx-mn)
x
```

```python
w=[1]*7
b=0
pred=np.dot(x,w)
pred=pred+b
pred=1/(1+np.exp(-pred))
print(pred)
print('---')
loss=np.multiply(-y, np.log(pred))-np.multiply((1-y),np.log(pred))
print(loss)
loss=loss.sum()/891
print(loss)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, shuffle=True)
print(x_train)
print(y_train)
```

```python
import matplotlib.pyplot as plt
cost=[]
epochs=[]
w=[1]*7
b=0
w_temp=[0]*7
b_temp=0
for i in range(1000):
    pred=np.dot(x_train,w)
    pred=pred+b
    pred=1/(1+np.exp(-pred))
    loss=np.multiply(y_train, np.log(pred))+np.multiply((1-y_train),np.log(1-pred))
    loss=-loss.sum()/668
    error=pred-y_train
    error=error.dot(error)
    error=error/(2*668)
    cost.append(loss)
    epochs.append(i)
    print(loss, error)
    for j in range(7): 
        w_temp[j]=w[j]-0.1/668*np.dot((pred-y_train),(x_train.iloc[:,j]))-0.1/668*w[j]# regularization parameter=1
    b_temp=b-0.1/668*(pred-y_train).sum() 
    w=w_temp
    b=b_temp
```

```python
plt.plot(epochs, cost)
plt.show()
```

```python
print('w= ',w)
print('b= ',b)
pred_y=np.dot(x_test, w)
pred_y=pred_y+b
pred_y=1/(1+np.exp(-pred_y))
loss=np.multiply(y_test, np.log(pred_y))+np.multiply((1-y_test),np.log(1-pred_y))
loss=-loss.sum()/223
loss
```
